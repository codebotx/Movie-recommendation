```python
movmat=df.pivot_table(index='user_id',columns='title',values='rating')
```

```python
movmat.head()
```

```python
ratings.sort_values('num_ratings',ascending=False).head()
```

```python
startwars_userratings=movmat['Star Wars (1977)']
```

```python
startwars_userratings.head()
```

```python
similar_to_starwars=movmat.corrwith(startwars_userratings)
```

```python
corr_starwars=pd.DataFrame(similar_to_starwars, columns=['Correlation'])
```

```python
corr_starwars.dropna(inplace=True)
```

```python
corr_starwars
```

```python
corr_starwars.sort_values('Correlation',ascending=False).head(10)
```

```python
corr_starwars=corr_starwars.join(ratings['num_ratings'])
```

```python
corr_starwars.head()
```

```python
corr_starwars[corr_starwars['num_ratings']>100].sort_values('Correlation', ascending=False)
```

```python
##Prediction
```

```python
def predict_moviename(movie_name):
    movie_userratings=movmat[movie_name]
    similar_to_moviename=movmat.corrwith(movie_userratings)

    corr_moviename=pd.DataFrame(similar_to_moviename, columns=['Correlation'])
    corr_moviename.dropna(inplace=True)

    corr_moviename=corr_moviename.join(ratings['num_ratings'])

    predictions=corr_moviename[corr_moviename['num_ratings']>100].sort_values('Correlation', ascending=False)

    return predictions
```

```python
predict_moviename("Titanic (1997)")
```
